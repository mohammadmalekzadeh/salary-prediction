### Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read dataframe

```python
df = pd.read_csv('../data/train_clean_data.csv')
df.head()
```

### Set default style for seaborn

```python
sns.set(style="whitegrid")
```

### Distribution of Annual Salary

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Annual_salary'], bins=30, kde=True)
plt.title('Distribution of Annual Salary')
plt.xlabel('Annual Salary')
plt.ylabel('Frequency')
plt.savefig('../reports/chart/Distribution of Annual Salary.png')
plt.show()
```

### Boxplot of Annual Salary by Job Title

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Job_title', y='Annual_salary', data=df)
plt.xticks(rotation=45)
plt.title('Salary Distribution by Job Title')
plt.tight_layout()
plt.savefig('../reports/chart/Salary Distribution by Job Title.png')
plt.show()
```

### Boxplot of Annual Salary by Department

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Annual_salary', data=df)
plt.xticks(rotation=45)
plt.title('Salary Distribution by Department')
plt.tight_layout()
plt.savefig('../reports/chart/Salary Distribution by Department.png')
plt.show()
```

### Scatter plot: Age vs Annual Salary

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Annual_salary', hue='Gender', data=df)
plt.title('Age vs Annual Salary')
plt.xlabel('Age')
plt.ylabel('Annual Salary')
plt.legend()
plt.savefig('../reports/chart/Age vs Annual Salary.png')
plt.show()
```

### Violin plot of Annual Salary by Gender

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='Gender', y='Annual_salary', data=df)
plt.title('Salary Distribution by Gender')
plt.savefig('../reports/chart/Salary Distribution by Gender.png')
plt.show()
```

### Bar plot: mean salary by Country

```python
plt.figure(figsize=(10,6))
mean_salary_country = df.groupby('Country')['Annual_salary'].mean().sort_values(ascending=False)
sns.barplot(x=mean_salary_country.values, y=mean_salary_country.index)
plt.title('Average Salary by Country')
plt.xlabel('Average Salary')
plt.ylabel('Country')
plt.savefig('../reports/chart/Average Salary by Country.png')
plt.show()
```

### Count plot: Number of employees by City

```python
plt.figure(figsize=(12,8))
sns.countplot(y='City', data=df, order=df['City'].value_counts().index)
plt.title('Number of Employees per City')
plt.ylabel('City')
plt.xlabel('Count')
plt.savefig('../reports/chart/Number of Employees per City.png')
plt.show()
```

### Relationship between Work Duration and Salary

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Work_duration', y='Annual_salary', hue='Job_title', data=df)
plt.title('Work Duration vs Annual Salary')
plt.xlabel('Work Duration (years)')
plt.ylabel('Annual Salary')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('../reports/chart/Work Duration vs Annual Salary.png')
plt.show()
```
